```python
# prompt: split df so that we can train randomforest our target column is "Fertilizer Name"
# Identify Non-Numerical Columns in X: Check the data types of the columns in the X DataFrame after dropping the target column.
# One-Hot Encode Categorical Columns: Use pd.get_dummies() to convert these categorical columns into a numerical format (binary columns for each category).
# import random forest and comment its parameter so i can tune it easyly.
# also predict accuracy score  Mean Average Precision @ 3 (MAP@3):

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder

# Drop the target column to create the feature set X
X = df.drop('Fertilizer Name', axis=1)
y = df['Fertilizer Name']

# Identify non-numerical columns in X
non_numerical_cols = X.select_dtypes(exclude=np.number).columns.tolist()

print("Non-numerical columns:", non_numerical_cols)

# One-Hot Encode Categorical Columns
X_encoded = pd.get_dummies(X, columns=non_numerical_cols, drop_first=True) # drop_first=True to avoid multicollinearity

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42, stratify=y)

# Encode the target variable if it's not already numerical
# RandomForestClassifier works best with numerical target variables
le = LabelEncoder()
y_train_encoded = le.fit_transform(y_train)
y_test_encoded = le.transform(y_test) # Use the same encoder fitted on the training data

# Initialize the Random Forest Classifier
# Key parameters for tuning:
rf_classifier = RandomForestClassifier(
    n_estimators=300,         # The number of trees in the forest. Higher generally improves performance but increases computation time.
    criterion='gini',         # The function to measure the quality of a split. 'gini' or 'entropy'.
    max_depth=None,           # The maximum depth of the tree. None means nodes are expanded until all leaves are pure or until all leaves contain less than min_samples_split samples. Controls overfitting.
    min_samples_split=3,      # The minimum number of samples required to split an internal node. Can prevent overfitting.
    min_samples_leaf=2,       # The minimum number of samples required to be at a leaf node. Can prevent overfitting.
    min_weight_fraction_leaf=0.0, # The minimum weighted fraction of the sum total of weights of all samples required to be at a leaf node.
    max_features='sqrt',      # The number of features to consider when looking for the best split. 'sqrt' (or 'auto') is common. Prevents individual trees from relying too heavily on any single feature.
    max_leaf_nodes=None,      # Grow trees with max_leaf_nodes in best first fashion.
    min_impurity_decrease=0.2, # A node will be split if this split induces a decrease of the impurity greater than or equal to this value.
    bootstrap=True,           # Whether bootstrap samples are used when building trees.
    oob_score=False,          # Whether to use out-of-bag samples to estimate the generalization accuracy.
    n_jobs=-1,                # Number of jobs to run in parallel. -1 uses all available processors.
    random_state=42,          # Controls the randomness of the estimator. Useful for reproducibility.
    verbose=0,                # Controls the verbosity when fitting and predicting.
    warm_start=False,         # When set to True, reuse the solution of the previous call to fit and add more estimators to the ensemble, otherwise, just erase the previous solution.
    class_weight=None,        # Weights associated with classes. Useful for imbalanced datasets.
    ccp_alpha=0.0             # Complexity parameter used for Minimal Cost-Complexity Pruning.
)

# Train the classifier
rf_classifier.fit(X_train, y_train_encoded)

# Predict on the test data
y_pred_encoded = rf_classifier.predict(X_test)
y_pred = le.inverse_transform(y_pred_encoded) # Convert predictions back to original labels

# Calculate Accuracy Score
accuracy = accuracy_score(y_test_encoded, y_pred_encoded)
print(f"Accuracy Score: {accuracy:.4f}")

# Calculate Mean Average Precision @ 3 (MAP@3)
# MAP@K requires predicted probabilities and ranking.
# We'll get the top K predictions and calculate precision for each sample.
# This is more involved than simple accuracy. We need the probability estimates for each class.
from sklearn.metrics import average_precision_score

# Get probability estimates for each class
y_pred_proba = rf_classifier.predict_proba(X_test)

# Function to calculate average precision @ K for a single sample
def apk(actual, predicted, k=3):
    if not actual:
        return 0.0

    score = 0.0
    num_hits = 0.0
    actual_set = set(actual)

    for i, p in enumerate(predicted[:k]):
        if p in actual_set and p not in predicted[:i]:
            num_hits += 1.0
            score += num_hits / (i + 1.0)

    return score / min(len(actual_set), k) # Normalize by the number of relevant items or K, whichever is smaller

# Function to calculate Mean Average Precision @ K
def mapk(actuals, predicteds, k=3):
    return np.mean([apk(a, p, k) for a, p in zip(actuals, predicteds)])

# Get the top 3 predicted classes for each sample based on probabilities
# We need to map the encoded class indices back to original labels to compare with y_test
top_k_indices = np.argsort(-y_pred_proba, axis=1)[:, :3]
top_k_labels = [le.inverse_transform(indices) for indices in top_k_indices]

# Prepare the actual labels in a list of lists format (each inner list contains the single correct label)
# For multi-label scenarios, the inner list could have multiple items. Here, it's single-label.
actual_labels = [[label] for label in y_test]

# Calculate MAP@3
map_at_3 = mapk(actual_labels, top_k_labels, k=3)
print(f"Mean Average Precision @ 3 (MAP@3): {map_at_3:.4f}")
```
